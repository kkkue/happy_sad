```python
import cv2
import glob
```

```python
from icrawler.builtin import GoogleImageCrawler, BingImageCrawler

def scraype(dir_path, word, num):
    crawler = GoogleImageCrawler(storage={"root_dir": dir_path})
    crawler.crawl(keyword=word, max_num=num)

def scraype_test(dir_path, word, num):
    crawler = BingImageCrawler(storage={"root_dir": dir_path})
    crawler.crawl(keyword=word, max_num=num)
```

```python
n_data = 100
scraype('/content/drive/MyDrive/Colab Notebooks/vgg16/img/image/faces/happy', '笑顔', n_data)
```

```python
n_data = 50
scraype_test('/content/drive/MyDrive/Colab Notebooks/vgg16/img/image/faces_test/happy', '笑顔', n_data)
```

```python
# 顔に着目して切り抜き
HAAR_FILE = '/content/drive/MyDrive/Colab Notebooks/vgg16/haarcascade_frontalface_alt.xml'
cascade = cv2.CascadeClassifier(HAAR_FILE)
```

```python

load_img_list = []
face_cut = []
count= 0
#出力画像のディレクトリパス
output_path = '/content/drive/MyDrive/Colab Notebooks/vgg16/img/image/train/happy/'
#加工元の画像ディレクトリパス
img_list = glob.glob('/content/drive/MyDrive/Colab Notebooks/vgg16/img/image/faces/happy/*.jpg')

for img in img_list:
    load_img_list.append(cv2.imread(img))
 
for load_img in load_img_list:
    face_rects = cascade.detectMultiScale(load_img)
    print(face_rects)
    if(len(face_rects) !=0):
        for face_rect in face_rects:
            x = face_rect[0]
            y = face_rect[1]
            w = face_rect[2]
            h = face_rect[3]
            face_cut = load_img[y:y + h, x:x + w]
            face_cut = cv2.resize(face_cut, dsize =(256, 256))
            img_name = output_path + "img_" + str(count) + ".jpg"
            cv2.imwrite(img_name, face_cut)
            count += 1
    else:
      pass
```
